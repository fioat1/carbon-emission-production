# 特征筛选

lasso

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Lasso
from sklearn.linear_model import LassoCV
import warnings
warnings.filterwarnings('ignore')
```

```python
# 先对特征进行标准化
y = df['carbon（y）']
features_cols = df.columns[2:]
features = df[features_cols]

scaler = StandardScaler()
features_standar = scaler.fit_transform(features)
```

```python
# lasso
alphas = np.logspace(-4, 1, 50)
lassoCV = LassoCV(alphas=alphas, cv=20, max_iter=1000).fit(features_standar,y)
```

```python
mask = lassoCV.coef_ != 0.0
features_cols[mask]
```

```python
from matplotlib.pyplot import MultipleLocator

plt.rcParams['font.sans-serif']=['Microsoft YaHei']

MSEs = lassoCV.mse_path_
MSEs_mean = np.apply_along_axis(np.mean,1,MSEs)   #计算MSE
MSEs_std = np.apply_along_axis(np.std,1,MSEs)  #计算标准差
plt.figure(dpi = 1000) 
plt.errorbar(lassoCV.alphas_, MSEs_mean, yerr=MSEs_std, fmt="o" ,ms=2.5, mfc="r", mec="r", ecolor="lightblue", elinewidth=1 , capsize=3, capthick=1) 
plt.semilogx()
plt.axvline(lassoCV.alpha_,color = 'black', ls="--")
plt.xlabel('Lambda')
plt.ylabel('MSE')
ax=plt.gca()
Y_major_locator=MultipleLocator(0.05) 
#ax.yaxis.set_major_locator (Y_major_locator)
%config InlineBackend.figure_format='svg'
plt.show()
```

```python
#绘制每个特征随着lambda取值的变化图
#使用坐标下降计算 Lasso 路径：
coefs = lassoCV.path(features_standar, y, alphas=alphas, max_iter = 1000)[1].T  #[1]代表从第2列开始（第一个特征），T代表矩阵转置

plt.figure(figsize=(10,5))
plt.semilogx(lassoCV.alphas_, coefs, '-') #x取对数坐标
plt.axvline(lassoCV.alpha_,color = 'black',ls="--")
plt.xlabel('Lambda')
plt.ylabel('Coefficients')
%config InlineBackend.figure_format='svg'
plt.legend(features_cols, frameon=False ,loc=1, fontsize='x-small', ncol=2)
plt.show ()
```

# 模型构建

```python
X_cols = ['煤炭消费量', '平板玻璃产量', '生铁产量', '水泥产量', '粗钢产量', '能源消费量', '第二产业结构', '天然气消费量']
X = df[X_cols]
y = df['carbon（y）']
```

```python
from sklearn import svm
```

```python
def fitness(x):
    # 计算适应度函数
    return np.sum(x ** 2)

def diff_gwo_svm(X, y):
    # 初始化参数
    dim = X.shape[1]
    pop_size = 30
    max_iter = 100
    lb = -1
    ub = 1

    # 初始化种群
    X_pos = np.random.uniform(lb, ub, (pop_size, dim))
    X_alpha = np.random.uniform(lb, ub, (pop_size, dim))
    X_beta = np.random.uniform(lb, ub, (pop_size, dim))
    
    # 开始迭代
    for i in range(max_iter):
        # 计算适应度函数值
        fitness_pos = fitness(X_pos)
        fitness_alpha = fitness(X_alpha)
        fitness_beta = fitness(X_beta)

        # 更新alpha和beta值
        alpha_index = np.argmin(fitness_alpha)
        beta_index = np.argmin(fitness_beta)
        X_alpha_new = X_alpha[alpha_index] + 0.5 * (X_pos[beta_index] - X_alpha[beta_index])
        X_beta_new = X_beta[beta_index] + 0.5 * (X_pos[alpha_index] - X_beta[beta_index])
        
        # 更新位置
        r1 = np.random.rand(pop_size, dim)
        r2 = np.random.rand(pop_size, dim)
        A1 = 2 * r1 - 1
        A2 = 2 * r2 - 1
        C1 = 2 * np.random.rand(pop_size, dim)
        C2 = 2 * np.random.rand(pop_size, dim)
        D_alpha = abs(C1 * X_alpha_new - X_pos)
        D_beta = abs(C2 * X_beta_new - X_pos)
        X_pos_new = X_alpha_new - A1 * D_alpha + A2 * D_beta

        # 边界处理
        X_pos_new[X_pos_new < lb] = lb
        X_pos_new[X_pos_new > ub] = ub

        # 训练SVM模型并预测结果
        clf = svm.SVR(kernel='rbf', C=1e3, gamma=0.1)
        clf.fit(X_pos_new, y)
        y_pred = clf.predict(X_pos_new)
        
        # 更新alpha和beta值
        for j in range(pop_size):
            if fitness(X_pos_new[j]) < fitness_alpha[j]:
                X_alpha[j] = X_pos_new[j]
            elif fitness(X_pos_new[j]) < fitness_beta[j]:
                X_beta[j] = X_pos_new[j]

            if fitness(X_pos_new[j]) > fitness_pos[j]:
                X_pos[j] = (X_alpha[j] + X_beta[j]) / 2

    return clf
```
